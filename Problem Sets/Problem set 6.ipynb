```python
from scipy.linalg import hilbert
import numpy as np
from sympy import *
```

```python
for i in range(10):
    matrix = hilbert(i+1)
    det = np.linalg.det(matrix)
    print('Det of hilbert hilber matrix (' + str(i+1) + ') \n',matrix)
    print('determinat = ', det)
```
