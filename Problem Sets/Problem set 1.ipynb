```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["figure.figsize"] = (10,5)
plt.rcParams['axes.grid'] = True 
```

```python
# Chapter 2 : 2.1 - 30

A = np.array([[0.8, 0.3],[0.2, 0.7]])
u = np.array([1, 0])
v = np.array([0,1])

plt.figure()
for i in range(40):
    u = np.dot(A, u)   
    plt.plot(u, 'o')
plt.show()    
        
plt.figure()
for i in range(40):
    v = np.dot(A, v)   
    plt.plot(v, 'o')
plt.show()
```
